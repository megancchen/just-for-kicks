```python
import numpy as np
```

```python
# defining a function for a weighted coin

def flip():
    '''
    this function tosses a weighted coin
    'decision 1' occurs with a probability of 0.85, and 'decision 2' occurs with 0.15.
    '''
    # n = number of tosses, p = probability of success (decision 2)
    n, p = 1, 0.15
    
    # running a binomial function that takes in these inputs
    toss = np.random.binomial(n, p, 1)
    
    # if 1, return 'decision 2'
    if toss[0] == 1:
        return 'decision 2'
    else:
        return 'decision 1'
```

```python
flip()
```
